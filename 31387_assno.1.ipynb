5.Data normalization

```python
df_min_max_scaled = cs.copy()
  
# apply normalization techniques
for column in df_min_max_scaled.columns:
    df_min_max_scaled['Price'] = (df_min_max_scaled['Price'] - df_min_max_scaled['Price'].min()) / (df_min_max_scaled['Price'].max() - df_min_max_scaled['Price'].min())    
  
# view normalized data
print(df_min_max_scaled)
```

6.converting categorical variable into quantatitive variables

```python
column_names = ['Address','Rooms','Type','Price','Method','SellerG','Date','Distance','Postcode','Bedroom2','Bathroom','Car','Landsize','BuildingArea','YearBuilt','CouncilArea','Lattitude','Longtitude','Regionname','Propertycount']



new_cs=cs.copy()#use to copy data

# calling the get_dummies method the first parameter mentions the the name of the data frame to store the new data frame in
# the second parameter is the list of columns which if not mentioned returns the dummies for all categorical columns

new_cs = pd.get_dummies(new_cs, columns = column_names)#used to convert catogorical variable into dummy variable
print(new_cs.head())
```
